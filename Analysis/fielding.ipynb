```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("ipl_clean_data.csv")
```

```python
catches_df = df[df["wicket_kind"].isin(["caught"])]
```

```python
top_catchers = catches_df["fielders"].value_counts()
```

```python
all_keepers = df[df["wicket_kind"]=="stumped"]["fielders"].unique()
```

```python
most_catches = df[(df["wicket_kind"]=="caught") & (~df["fielders"].isin(all_keepers))]
```

```python
top_catchers = most_catches["fielders"].value_counts()
top_catchers = top_catchers.reset_index()
top_catchers.to_csv("top_catch_takers.csv",index=False)
```

```python
top_catchers.head(10)
```

```python
# Filter for stumpings
stumpings_df = df[df['wicket_kind'] == 'stumped']

# Count by fielder
most_stumpings = stumpings_df.groupby('fielders').size().sort_values(ascending=False)

# Reset index and save
most_stumpings = most_stumpings.reset_index()
most_stumpings.columns = ['wicketkeeper', 'stumpings']
most_stumpings.to_csv("most_stumpings.csv", index=False)

most_stumpings
```

```python
# 1. Saare keepers ki list nikaalo (Stumpings ke base par)
all_keepers = df[df["wicket_kind"] == "stumped"]["fielders"].unique()

# 2. Filter: Sirf wo catches jo keepers ne liye hain
caught_behind_df = df[(df["wicket_kind"] == "caught") & (df["fielders"].isin(all_keepers))]

# 3. Count aur Save
top_caught_behind = caught_behind_df["fielders"].value_counts().reset_index()
top_caught_behind.columns = ["wicket_keeper", "caught_behind_count"]
top_caught_behind.to_csv("most_caught_behind.csv", index=False)

top_caught_behind.head(10)
```

```python
runout_df = df[df["wicket_kind"]=="run out"]
top_runouts = runout_df["fielders"].value_counts().reset_index()
top_runouts.columns = ["fielder", "runout_counts"]
top_runouts.to_csv("most_runouts.csv",index=False)
top_runouts.head(10)
```
