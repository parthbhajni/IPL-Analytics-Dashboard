```python
# 1. Jo tune groupby kiya hai, uske aage .reset_index() laga de
top_runs_df = df.groupby('batter')['runs_batter'].sum().reset_index()

# 2. Columns ke naam thik kar de taaki CSV me saaf dikhe
top_runs_df.columns = ['player_name', 'total_runs']

# 3. Ab sort kar aur top 10 nikaal
top_runs_final = top_runs_df.sort_values(by='total_runs', ascending=False)

# 4. Ab ise save kar, ab player ka naam pakka aayega CSV mein
top_runs_final.to_csv('top_run_scorers.csv', index=False)
```

```python
balls_faced = df.groupby("batter").size()
```

```python
runs = df.groupby("batter")["runs_batter"].sum()
```

```python
strike_rate = (runs/balls_faced)*100
```

```python
sr_df = pd.DataFrame({
    "runs" : runs,
"balls" : balls_faced,
"strike_rate" : strike_rate})
```

```python
sr_df = sr_df[sr_df["balls"] > 1100]
```

```python
# 1. First, sort and take the top 10
sr_df = sr_df.sort_values("strike_rate", ascending=False)

# 2. Convert 'batter' from index to a column (This fixes the missing names)
sr_df = sr_df.reset_index()

# 3. Round the strike_rate to 2 decimal places
sr_df["strike_rate"] = sr_df["strike_rate"].round(2)

# 4. Save to CSV
sr_df.to_csv("death_strike_rate.csv", index=False)
```

```python
death_df = df[df["over"]>=16]
death_runs = death_df.groupby("batter")["runs_batter"].sum()
death_runs.sort_values(ascending = False)
```

```python
death_balls = death_df.groupby("batter").size()
death_sr = (death_runs/death_balls)*100
death_stats = pd.DataFrame({
    "runs": death_runs,
    "balls": death_balls,
    "strike_rate": death_sr
})

# 1. Filter
death_stats = death_stats[death_stats["balls"] >= 200]

# 2. SORT AND SAVE (Add 'death_stats =' at the start)
death_stats = death_stats.sort_values("runs", ascending=False)

# 3. RESET INDEX AND SAVE (Add 'death_stats =' at the start)
death_stats = death_stats.reset_index()

# 4. Round off
death_stats["strike_rate"] = death_stats["strike_rate"].round(2)

# 5. Save to CSV
death_stats.to_csv("death_statistics.csv", index=False)

# Check the output
death_stats.head(10)
```

```python
# 1. Calculate runs from 4s and 6s for each ball
df['boundary_runs'] = df.apply(lambda x: x['runs_batter'] if x['runs_batter'] in [4, 6] else 0, axis=1)

# 2. Group by batter to get total runs and boundary runs
batter_stats = df.groupby('batter').agg(
    total_runs=('runs_batter', 'sum'),
    total_boundary_runs=('boundary_runs', 'sum')
).reset_index()

# 3. Calculate the Percentage
batter_stats['boundary_percentage'] = (batter_stats['total_boundary_runs'] / batter_stats['total_runs']) * 100

# 4. Filter for players with significant career runs (e.g., > 500) to keep it relevant
top_boundary_hitters = batter_stats[batter_stats['total_runs'] > 1000].sort_values(by='boundary_percentage', ascending=False)

top_boundary_hitters[['batter', 'total_runs', 'boundary_percentage']].head(10)
```

```python
df["is_four"] = df["runs_batter"].isin([4])
df["is_six"] = df["runs_batter"].isin([6])

df["is_boundary"] = df["runs_batter"].isin ([4,6])

boundary_stats = df.groupby("batter").agg(
    fours_count = ("is_four","sum"),
    six_count=("is_six","sum"),
    boundary_count = ("is_boundary","sum"),
    boundary_runs_sum = ("boundary_runs","sum"),
    total_runs = ("runs_batter","sum")).reset_index()

top_hitters = boundary_stats.sort_values(by='total_runs',ascending = False)

top_hitters.head(10)
```
